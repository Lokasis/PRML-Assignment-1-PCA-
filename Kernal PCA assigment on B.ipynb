```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plot
```

```python
%matplotlib inline
```

```python
header_names=['first','second']
data=pd.read_csv("E:/competitive coding/zzzzz/PRML_assignment1/PRML_assignment1_2020/Dataset3.csv",header=None,names=header_names)
```

```python
data_arr=np.array(data)
```

```python
n=1000
temp=np.ones((n,n))/n
```

```python
sigma1=0.1
```

```python
kernal1 = np.zeros((1000,1000))
for i,x in enumerate(data_arr):
    for j,y in enumerate(data_arr):
        kernal1[i,j] = np.exp(np.linalg.norm(x-y,axis=0)**2/-(2*sigma1*sigma1))
```

```python
kernal1 = kernal1 - temp.dot(kernal1) - kernal1.dot(temp) + temp.dot(kernal1).dot(temp)
```

```python
eigenvalues1, eigenvectors1=np.linalg.eig(kernal1)
```

```python
indices1 = np.argsort(eigenvalues1)[::-1] 
eigenvalues1 = eigenvalues1[indices1]
eigenvectors1 = eigenvectors1[:,indices1]
```

```python
components1=2
matrix1=[]
for i in range (components1):
    matrix1.append(eigenvectors1[:,i])
matrix1=np.array(matrix1)
matrix1=np.array(matrix1.T)
```

```python
plot.figure(figsize=(8,6))
plot.scatter(matrix1[:,0],matrix1[:,1],color='g')
plot.grid(True)
plot.title("Gaussian Kernal PCA algorithm with sigma=0.1")
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.show()
```
