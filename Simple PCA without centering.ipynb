```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from matplotlib import rcParams
%matplotlib inline
```

```python
header_names=['first','second']
data=pd.read_csv("E:/competitive coding/zzzzz/PRML_assignment1/PRML_assignment1_2020/Dataset3.csv",header=None,names=header_names)
```

```python
plot.figure()
rcParams['figure.figsize'] = 6,4
plot.scatter(data['first'],data['second'])
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.title("Data before centering")
```

```python
data_arr=np.array(data)
```

```python
covariance_matrix=np.cov(data_arr.T)
```

```python
covariance_matrix=np.array(covariance_matrix)
```

```python
eigenvalues,eigenvectors=np.linalg.eig(covariance_matrix)
print(eigenvalues)
eigenvectors=np.array(eigenvectors)
print(eigenvectors.shape)
```

```python
indices = np.argsort(eigenvalues)[::-1] 
eigenvalues = eigenvalues[indices]
eigenvectors = eigenvectors[:,indices]
print(eigenvalues)
print(eigenvectors)
```

```python
eigenpair = [[eigenvectors[0][0]],
             [eigenvectors[0][1]]]
eigenpair=np.array(eigenpair)
print(eigenpair)
print(eigenpair.shape)
print(type(eigenpair))
print(eigenpair.shape)
print(data_arr.shape)
res=np.dot(data_arr,eigenpair)
res=np.array(res)
print(res.shape)
```

```python
plot.figure(figsize=(8,6))
plot.scatter(res*eigenpair[0],res*eigenpair[1],color='g', s=15, alpha=0.5)
plot.scatter(data['first'],data['second'])
plot.grid(True)
plot.title("Data projected on Largest Eigen Vector")
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.show()
```
